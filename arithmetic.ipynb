```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

train = pd.read_csv('mnist_train.csv')
test = pd.read_csv('mnist_test.csv')

rownum = 1500
if rownum > -1 and rownum < len(train):
    pixels = train.iloc[rownum][1:].values.reshape(28, 28)
    array = np.array(pixels, dtype=np.uint8)
    plt.imshow(array, cmap='gray')
    plt.title(f"MNIST Digit at Row {rownum}")
    plt.axis('off')
    plt.show()

x, y, w, h = 10, 10, 18, 18
cropped = array[y:y+h, x:x+w]
plt.imshow(cropped, cmap='gray')
plt.title("Cropped Digit")
plt.axis('off')
plt.show()

rownum_b = 16
if rownum_b > -1 and rownum_b < len(train):
    pixels_b = train.iloc[rownum_b][1:].values.reshape(28, 28)
    array_b = np.array(pixels_b, dtype=np.uint8)
    added_image = np.clip(array + array_b, 0, 255)
    subtracted_image = np.clip(array - array_b, 0, 255)
    plt.subplot(1, 3, 1)
    plt.imshow(array, cmap='gray')
    plt.title("Image A")
    plt.axis('off')
    plt.subplot(1, 3, 2)
    plt.imshow(array_b, cmap='gray')
    plt.title("Image B")
    plt.axis('off')
    plt.subplot(1, 3, 3)
    plt.imshow(added_image, cmap='gray')
    plt.title("Added Image")
    plt.axis('off')
    plt.show()
    plt.imshow(subtracted_image, cmap='gray')
    plt.title("Subtracted Image")
    plt.axis('off')
    plt.show()

bitwise_and = np.bitwise_and(array, array_b)
bitwise_or = np.bitwise_or(array, array_b)
bitwise_xor = np.bitwise_xor(array, array_b)
bitwise_not = np.bitwise_not(array, array_b)
plt.subplot(1, 4, 1)
plt.imshow(bitwise_and, cmap='gray')
plt.title("Bitwise AND")
plt.axis('off')
plt.subplot(1, 4, 2)
plt.imshow(bitwise_or, cmap='gray')
plt.title("Bitwise OR")
plt.axis('off')
plt.subplot(1, 4, 3)
plt.imshow(bitwise_xor, cmap='gray')
plt.title("Bitwise XOR")
plt.axis('off')
plt.subplot(1, 4, 4)
plt.imshow(bitwise_not, cmap='gray')
plt.title("Bitwise NOT")
plt.axis('off')
plt.show()

left_shifted = np.left_shift(array, 1)
right_shifted = np.right_shift(array, 1)
plt.subplot(1, 3, 1)
plt.imshow(array, cmap='gray')
plt.title("Original")
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(left_shifted, cmap='gray')
plt.title("Left Shift")
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(right_shifted, cmap='gray')
plt.title("Right Shift")
plt.axis('off')
plt.show()
```
